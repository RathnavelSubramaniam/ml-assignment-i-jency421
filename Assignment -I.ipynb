b) What is the range of each quantitative predictor? You can answer this using the range() function. 

```python
col = ['mpg','cylinders','displacement','horsepower','weight','acceleration','year']
min = auto[col].min()
max = auto[col].max()
print(f"Range of {col}: {min} , {max}")
```

* Interpretation

The output shows the minimum and maximum values for each selected quantitative variable:

 The range (max - min) for each variable tells us how widely the values are spread.
 For example:
   * If mpg ranges from 9 to 46.6, it means fuel efficiency varies significantly across vehicles.
   * If horsepower ranges from 46 to 230, it shows there's a large variation in engine power.
   * If year ranges from 70 to 82, the data spans model years from 1970 to 1982.

Understanding the range helps in:
Identifying potential outliers or extreme values.
Knowing the scale of each variable for further analysis or normalization.


C) What is the mean and standard deviation of each quantitative predictor? 

```python
mean = auto[col].mean()
std = auto[col].std()
print(f"{col}: {mean} , Std Dev of {col}: {std}")
```

* Interpretation

for the variable col in the full dataset:

* The mean represents the average value of all observations. It gives a central tendency of the data.
* The standard deviation (std)** indicates how spread out the values are from the mean.
   * A low std means values are close to the mean (less variation).
   * A high std means the values are more spread out (more variation).

These statistics help summarize the distribution of the predictor and understand how consistent or varied the data .


d) Now remove the 10th through 85th observations. What is the range, mean, and standard deviation of each predictor in the subset of the data that remains? 

```python
auto_subset = auto.drop(auto.index[9:85])
min = auto_subset[col].min()
max = auto_subset[col].max()
mean = auto_subset[col].mean()
std = auto_subset[col].std()
print(f"Range of {col}: {min} , {max}")
print(f"Mean of {col}: {mean} , Std Dev of {col}: {std}")
```

* Interpretation

After removing rows 10 through 85 from the dataset, we calculate the statistical summary for the predictor variable `{col}`:

* The range (difference between min and max) gives an idea of the spread of values.
* The mean indicates the average value of the variable.
* The standard deviation tells us how much the values deviate from the mean.

Comparing these values with the original full dataset can help us understand how this subset (with fewer rows) changes the distribution.

or example, if the mean or standard deviation changes significantly, it means those dropped observations had a strong influence on the data.


e) Using the full data set, investigate the predictors graphically, using scatterplots or other tools of your choice. Create some plots highlighting the relationships among the predictors. Comment on your findings.

```python


sns.pairplot(auto, diag_kind='hist')
plt.suptitle("Pairplot of Quantitative Variables", y=1.02)
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(auto.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Quantitative Variables")
plt.show()


sns.boxplot(x='cylinders', y='mpg', data=auto)
plt.title("MPG Distribution by Number of Cylinders")
plt.show()


sns.lmplot(x='weight', y='mpg', data=auto)
plt.title("MPG vs Weight with Regression Line")
plt.show()

sns.scatterplot(auto[col])
plt.suptitle("Scatterplot of Auto Data")
plt.show()
```

* Interpretation

* The pairplot shows strong negative linear relationships between mpg and variables like weight, horsepower, and displacement.

* The heatmap confirms this â€” with mpg having strong negative correlations with:
   - weight (-0.83)
   - horsepower (-0.78)
   - displacement (-0.81)

* The boxplot shows that cars with more cylinders tend to have lower mpg.

* The regression plot between weight and mpg shows a clear negative trend, reinforcing that heavier cars are less fuel-efficient.

Overall, the graphical analysis suggests strong relationships among the predictors, especially those linked to engine size and vehicle weight.


f)Suppose that we wish to predict gas mileage (mpg) on the basis of the other variables. Do your plots suggest that any of the other variables might be useful in predicting mpg? Justify your answer

```python
predictors = ['horsepower', 'weight', 'displacement', 'acceleration', 'cylinders']
sns.pairplot(auto, x_vars=predictors, y_vars='mpg', height=4, kind='scatter')
plt.suptitle("Scatterplots: mpg vs Other Predictors", y=1.02)
plt.show()
```

```python
correlations = auto.corr(numeric_only=True)['mpg'].sort_values()
print("Correlation of mpg with other variables:\n")
print(correlations)
```

* Interpretation


From the scatterplots and correlation values:

- 'horsepower', 'weight', and 'displacement' show strong negative correlation with 'mpg'.
- 'weight' especially shows a very clear downward trend â€” heavier cars tend to have lower mileage.
- 'acceleration' and 'cylinders' have weaker relationships with 'mpg'.

Therefore, 'horsepower', 'weight', and 'displacement' are the most useful predictors for estimating mpg.
